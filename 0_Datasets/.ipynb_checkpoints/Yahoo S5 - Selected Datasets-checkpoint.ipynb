# Showcase of selected Yahoo S5 Time Series

This jupyter notebook is used for visualizing the Yahoo S5 datasets which have been selected for evaluating the performance of anomaly detection algorithms.

## Import Packages and define Function Prototypes

```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
def plotTimeSeries(filePath, selectedTimeSeries, fileNamePrefix):
    """    
        load datasets and create pandas dataframes
    """
    dataFiles = []
    for i in range(len(selectedTimeSeries)):
        dataFiles.append(fileNamePrefix + str(selectedTimeSeries[i]) + '.csv')

    datasets = []
    fileNames = []
    counter = 0
    for file in dataFiles:
        if 'csv' in file:
            counter +=1
            raw_data = pd.read_csv(os.path.join(filePath,file))
            df = raw_data
            df = df.sort_values('timestamp')
            datasets.append(df)
            fileNames.append(file)

    """    
        plot choosen datasets with anomaly ground truth
    """        
    fig, ax = plt.subplots(nrows = counter, figsize=(20,4*counter), constrained_layout=True)
    fig.set_constrained_layout_pads(w_pad=0.1, h_pad=0.2, hspace=0, wspace=0)

    for i in range(counter):

        ax[i].plot(datasets[i]['timestamp'], datasets[i]['value'],label ='Time Series Data')
        #ax[i].plot(datasets[i].index.values, datasets[i]['value'],label ='Normal')
        ax[i].set(xlabel="Index (Timestamp)", ylabel="Signal Value", title='Time Series Name: "'+ str(fileNames[i]) +'"')
        ax[i].grid()

        datasetLength = len(datasets[i])

        for j in range(datasetLength):
            if datasets[i]['is_anomaly'][j] == 1:
                x = datasets[i]['timestamp'][j]
                y = datasets[i]['value'][j]
                ax[i].scatter(x,y,color='red', s=50)

        legendElements = [Line2D([0], [0], color='C0', lw=2, label='Time Series Data'),
                          Line2D([0], [0], marker='o', color='w', label='Anomaly Ground Truth', markerfacecolor='red', markersize=8)]
        ax[i].legend(handles=legendElements)

    plt.show()
```

# Yahoo S5-A1 - Real Data of Yahoo Dataset 

<blockquote>
Note that the timestamps of the A1Benchmark are replaced by integers with the increment of 1, where each data-point represents 1 hour worth of data. The anomalies in A1Benchmark are marked by humans and therefore may not be consistent, therefore during benchmarking A1Benchmark is best used to measure recall.

The fields are:
- 0 : timestamp
- 1 : value
- 2 : is_anomaly

The is_anomaly field is a boolean indicating if the current value at a given timestamp is considered an anomaly.
</blockquote>

```python
A1filePath = 'Yahoo-S5\\A1Benchmark'

A1SelectedDatasets =[9,
                    15,
                    17,
                    23,
                    34,
                    39,
                    46,
                    52,
                    53,
                    62]

A1FileNamePrefix = 'real_'

plotTimeSeries(A1filePath, A1SelectedDatasets, A1FileNamePrefix)
```

# Yahoo S5-A2 - Synthetic Data of Yahoo Dataset

<blockquote>
The synthetic data set contains time-series with random seasonality, trend and noise. The outliers in the synthetic dataset are inserted at random positions.

The fields are:
- 0 : timestamp
- 1 : value
- 2 : is_anomaly

The is_anomaly field is a boolean indicating if the current value at a given timestamp is considered an anomaly.
</blockquote>

```python
A2filePath = 'Yahoo-S5\\A2Benchmark'

A2SelectedDatasets =[1,
                    8,
                    10,
                    17,
                    22,
                    23,
                    26,
                    29,
                    32,
                    33]

A2FileNamePrefix = 'synthetic_'

plotTimeSeries(A2filePath, A2SelectedDatasets, A2FileNamePrefix)
```
