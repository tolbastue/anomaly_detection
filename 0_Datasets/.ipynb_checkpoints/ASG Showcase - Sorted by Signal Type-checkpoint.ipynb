# Showcase of the Artificial Signal Generator (ASG) Dataset
## (sorted by base signal types)

This jupyter notebook is used to visualize all ASG (Artificial Signal Generator) time series.

The Artificial Signal Generator (ASG) is a personally created dataset, which is designed for the evaluation of anomaly detection algorithms for time series data. The dataset contains in total 18 time series which have been generated by implementing a function generator in CODESYS and sampling signal values at 100 Hz using a industrial PLC. 

The ASG dataset contains the following base signal types:
- sine wave,
- rectangular wave (square signal) and
- triangular wave (ramp signal).

Each base signal is used for creating an anomalous signal by using the following anomaly types:
- point anomalies,
- collective anomalies ("Hold" anomalies in further context) and
- contextual anomalies ("MeanChange" anomalies in further context).

## Import Packages and define Function Prototypes

```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
def plotTimeSeries(datasetPath):
    filePath,_,datasetFiles = next(os.walk(datasetPath))

    datasets = []
    fileNames = []
    fileCounter = 0
    for file in datasetFiles:
        if 'csv' in file:
            fileCounter += 1
            raw_data = pd.read_csv(os.path.join(datasetPath,file), delimiter=';')
            df = raw_data
            datasets.append(df)
            fileNames.append(file)

    fig, ax = plt.subplots(nrows = fileCounter,figsize=(20,4*fileCounter))
    fig.subplots_adjust(hspace = 0.5)

    for i in range(0,fileCounter):

        anomaly_groundTruth = datasets[i].loc[datasets[i]['is_anomaly'] == 1, ['timestamp', 'value']]

        ax[i].plot(datasets[i]['timestamp'], datasets[i]['value'],label ='time series data')
        ax[i].scatter(anomaly_groundTruth['timestamp'], anomaly_groundTruth['value'], color='red', label = 'anomaly ground truth', s = 50, marker = "o")
        ax[i].set(ylabel="value", xlabel="time (ms)", title='dataset name: "'+ str(fileNames[i])+'"')
        ax[i].grid()
        ax[i].legend()
```

## Visualize created Sine Datasets

```python
plotTimeSeries('ASG\Sine')
```

## Visualize created Rectangular Datasets

```python
plotTimeSeries('ASG\Rect')
```

## Visualize created Ramp Datasets

```python
plotTimeSeries('ASG\Ramp')
```
